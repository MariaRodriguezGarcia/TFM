```python
import pandas as pd
print(len(rows), len(folders))
print(len(rows_conn_log), len(folders_conn_log))

df = pd.DataFrame({'rows': rows, 'folders': folders})
df_logs = pd.DataFrame({'rows': rows_conn_log, 'folders': folders_conn_log})
df.head()
```

```python
# Sort the DataFrame by the 'folders' column
df_sorted = df.sort_values(by='folders')

# Display the sorted DataFrame
print(df_sorted)
```

```python
# Sort the DataFrame by the 'folders' column
df_logs_sorted = df_logs.sort_values(by='folders')

# Display the sorted DataFrame
print(df_logs_sorted)
```

```python
valor_falta = df_sorted[~df_sorted['folders'].isin(df_logs_sorted['folders'])]['folders']
print("Valor de df_sorted que no está en df_logs_sorted: ", valor_falta)
```

```python
df_sorted =df_sorted[~df_sorted["folders"].isin(["CTU-IoT-Malware-Capture-39-1"])]
print(len(df_sorted))
print(df_sorted)
```

```python
import matplotlib.pyplot as plt

# Define the width of each bar
bar_width = 0.35
import numpy as np

# Define the x positions for the bars
x1 = np.arange(len(df_sorted["folders"]))
x2 = [x + bar_width for x in x1]

# Create a figure and axis object with a larger size
fig, ax = plt.subplots(figsize=(12, 8))  # Adjust the size as needed

# Plot the histograms for each dataframe
ax.bar(x1, df_sorted["rows"], width=bar_width, color='blue', label='df_sorted')
ax.bar(x2, df_logs_sorted["rows"], width=bar_width, color='orange', label='df_logs_sorted')

# Set the x-axis tick positions and labels
ax.set_xticks([x + bar_width/2 for x in range(len(df_sorted["folders"]))])
ax.set_xticklabels(df_sorted["folders"], rotation=45, ha='right')  # Rotate by 45 degrees and align to the right

# Set y-axis scale to logarithmic
ax.set_yscale('log')

# Add labels and title
plt.xlabel('Folders')
plt.ylabel('Number of Rows')
plt.title('Comparison of Number of Rows by Folder')

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
