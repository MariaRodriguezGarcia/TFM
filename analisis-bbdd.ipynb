```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming df is your DataFrame
# Replace 'Duration' and 'Label' with the actual column names

# Group the DataFrame by 'Label'
grouped_df = df.groupby('label')

# Determine the layout of subplots
labels_per_plot = 8  # Number of labels per plot
num_plots = int(np.ceil(len(grouped_df) / labels_per_plot))

# Create and plot each subset of labels
for i in range(num_plots):
    start_idx = i * labels_per_plot
    end_idx = min((i + 1) * labels_per_plot, len(grouped_df))
    subset_labels = list(grouped_df.groups.keys())[start_idx:end_idx]
    
    # Create subplots with specified layout
    num_cols = 2  # Number of columns for subplots
    num_rows = int(np.ceil(len(subset_labels) / num_cols))  # Number of rows for subplots
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
    fig.subplots_adjust(hspace=0.5)  # Adjust vertical spacing between subplots
    
    # Flatten axes if there's only one row
    if num_rows == 1:
        axes = np.array([axes])
    
    # Plot histogram for each group in the subset
    for idx, label in enumerate(subset_labels):
        ax = axes[idx // num_cols, idx % num_cols]  # Get the current axis
        group = grouped_df.get_group(label)
        ax.hist(group['flow_duration'], bins=20, alpha=0.5)
        ax.set_title(label)
        ax.set_xlabel('flow_duration')
        ax.set_ylabel('Frequency')
    
    # Hide empty subplots
    for idx in range(len(subset_labels), num_rows * num_cols):
        ax = axes[idx // num_cols, idx % num_cols]  # Get the current axis
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()
```
